```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
import time

# --- Configuration ---
# Define the download folder. Make sure this folder exists!
DOWNLOAD_FOLDER = "/path/to/your/download/folder"  # **<<< CHANGE THIS TO A VALID PATH >>>**
# Example for Windows: "C:/Users/YourUser/Downloads"
# Example for Linux/Mac: "/home/youruser/downloads"

WEBSITE_URL = "https://iprsearch.ipindia.gov.in/PublicSearch"

# --- XPaths ---
# !! IMPORTANT: These might need adjustment based on the actual page structure !!
# XPath to find the table containing the application numbers (adjust if needed)
TABLE_XPATH = "//table" # Example: might be more specific like "//table[@id='resultsTable']"
# XPath to find the buttons within the first column of the table body
# Assumes the button is directly inside the first <td> of each <tr> in <tbody>
BUTTON_XPATH = "//table//tbody/tr/td[1]//button | //table//tbody/tr/td[1]//a" # Try button first, then link if button not found
# XPath for an element expected to be present on the page *after* clicking a button (adjust if needed)
# Using 'body' as a fallback, but a more specific element is better.
DETAIL_PAGE_ELEMENT_XPATH = "//body" # Example: might be "//div[@id='applicationDetails']" or similar

# --- Functions ---

def extract_text_from_website(driver):
    """
    Extracts all visible text from the current page the webdriver is on.

    Args:
        driver: The Selenium webdriver instance.

    Returns:
        A string containing all the extracted text, or None on error.
    """
    try:
        # Wait for a specific element unique to the detail page/view if possible
        # Or just wait for the body tag to ensure basic loading
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, DETAIL_PAGE_ELEMENT_XPATH))
        )
        # Add a small buffer wait
        time.sleep(1)
        # Extract the text from the body
        page_text = driver.find_element(By.TAG_NAME, 'body').text
        return page_text
    except TimeoutException:
        print("Error: Timed out waiting for detail page content to load.")
        return None
    except NoSuchElementException:
        print("Error: Could not find the body element to extract text.")
        return None
    except Exception as e:
        print(f"Error extracting text: {e}")
        return None

def setup_driver():
    """Sets up and returns the Selenium Chrome driver."""
    options = webdriver.ChromeOptions()
    # Ensure the download folder path is valid
    if DOWNLOAD_FOLDER == "/path/to/your/download/folder":
        print("WARNING: DOWNLOAD_FOLDER path is not set. Downloads might not work as expected.")
    else:
        options.add_experimental_option("prefs", {
            "download.default_directory": DOWNLOAD_FOLDER,
            "plugins.always_open_pdf_externally": True,
            "download.prompt_for_download": False
        })
    # Add arguments to potentially help stability
    options.add_argument("--start-maximized")
    options.add_argument("--disable-extensions")
    options.add_argument("--disable-gpu")
    # options.add_argument("--headless") # Uncomment for headless mode (no browser UI)

    try:
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
        return driver
    except Exception as e:
        print(f"Error initializing WebDriver: {e}")
        print("Please ensure Chrome is installed and accessible.")
        print("You might need to check Chromedriver compatibility or network connection.")
        return None

def main():
    """
    Opens the website, waits for login, finds application number buttons,
    clicks each, extracts text, and navigates back.
    """
    driver = setup_driver()
    if not driver:
        return # Exit if driver setup failed

    try:
        # Open the website
        driver.get(WEBSITE_URL)
        print(f"Opened {WEBSITE_URL}.")
        print(">>> Please log in manually in the browser window.")
        input(">>> After logging in and the results table is visible, press Enter here to continue...")

        print("Attempting to find and process application number buttons...")

        # --- Wait for the main table to be present after login ---
        try:
            WebDriverWait(driver, 30).until(
                EC.presence_of_element_located((By.XPATH, TABLE_XPATH))
            )
            print("Results table found.")
        except TimeoutException:
            print(f"Error: Timed out waiting for the results table ({TABLE_XPATH}) after login.")
            print("Please ensure you are logged in and the table is visible.")
            return # Exit if table not found

        # --- Get the initial count of buttons ---
        try:
            initial_buttons = driver.find_elements(By.XPATH, BUTTON_XPATH)
            num_buttons = len(initial_buttons)
            if num_buttons == 0:
                print(f"Error: No application buttons found using XPath: {BUTTON_XPATH}")
                print("Please check the BUTTON_XPATH selector against the page HTML.")
                return
            print(f"Found {num_buttons} application buttons/links to process.")
        except Exception as e:
            print(f"Error finding initial buttons: {e}")
            return

        # --- Loop through each button by index ---
        for i in range(num_buttons):
            print(f"\n--- Processing button {i + 1} of {num_buttons} ---")
            try:
                # **Crucial: Re-find the buttons in each iteration to avoid stale elements**
                WebDriverWait(driver, 20).until(
                     EC.presence_of_element_located((By.XPATH, TABLE_XPATH)) # Ensure table is present first
                )
                buttons = WebDriverWait(driver, 20).until(
                    EC.presence_of_all_elements_located((By.XPATH, BUTTON_XPATH))
                )

                # Sanity check: ensure button index is still valid
                if i >= len(buttons):
                     print(f"Warning: Button index {i} out of range after re-finding elements. Skipping remaining.")
                     break

                button = buttons[i]
                button_text = button.text.strip() # Get text for context
                print(f"Attempting to click button/link with text: '{button_text}'")

                # Scroll into view and click
                driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", button)
                time.sleep(0.5) # Small pause before click
                # Use JavaScript click as a fallback if direct click fails sometimes
                try:
                    WebDriverWait(driver, 10).until(EC.element_to_be_clickable(button)).click()
                except Exception as click_err:
                    print(f"Direct click failed ({click_err}), trying JavaScript click...")
                    driver.execute_script("arguments[0].click();", button)

                print("Button clicked. Waiting for new content...")

                # --- Extract text from the new view ---
                page_text = extract_text_from_website(driver)
                if page_text:
                    print(f"\n--- Extracted Text for '{button_text}' ---")
                    print(page_text[:1000] + "..." if len(page_text) > 1000 else page_text) # Print truncated text
                    print("--- End of Extracted Text ---")
                else:
                    print("Failed to extract text from the detail view.")

                # --- Navigate back ---
                print("Navigating back to the results list...")
                driver.back()

                # --- Wait for the results list page to reload ---
                print("Waiting for results list to reload...")
                WebDriverWait(driver, 20).until(
                    EC.presence_of_element_located((By.XPATH, TABLE_XPATH)) # Wait for table
                )
                WebDriverWait(driver, 20).until(
                    EC.presence_of_all_elements_located((By.XPATH, BUTTON_XPATH)) # Wait for buttons
                )
                time.sleep(1) # Extra buffer time after navigation back
                print("Results list reloaded.")

            except StaleElementReferenceException:
                print(f"Error: Stale element reference encountered for button {i + 1}. The page might have changed unexpectedly. Skipping this button.")
                # Attempt to recover by refreshing or just continue
                time.sleep(2) # Wait a bit before next iteration
                continue # Move to the next button
            except TimeoutException:
                print(f"Error: Timed out waiting for elements related to button {i + 1}. Skipping.")
                # Consider navigating back if stuck on a detail page
                current_url = driver.current_url
                if WEBSITE_URL not in current_url: # Basic check if not on main page
                    try:
                         print("Attempting to navigate back after timeout...")
                         driver.back()
                         WebDriverWait(driver, 10).until(EC.url_contains(WEBSITE_URL.split('/')[-1])) # Wait to be back roughly
                    except Exception as back_err:
                         print(f"Failed to navigate back after timeout: {back_err}")
                continue
            except Exception as e:
                print(f"An unexpected error occurred processing button {i + 1}: {e}")
                import traceback
                traceback.print_exc() # Print detailed traceback for debugging
                print("Attempting to continue with the next button...")
                # Try navigating back if possible
                try:
                    if WEBSITE_URL not in driver.current_url:
                        driver.back()
                        WebDriverWait(driver, 10).until(EC.url_contains(WEBSITE_URL.split('/')[-1]))
                except Exception:
                    pass # Ignore errors during recovery attempt
                continue # Move to the next button

        print("\nFinished processing all found buttons.")

    except Exception as e:
        print(f"\nAn unexpected error occurred in the main process: {e}")
        import traceback
        traceback.print_exc()
    finally:
        print("Closing the browser.")
        if driver:
            driver.quit() # Ensure the driver is closed

if __name__ == "__main__":
    main()
```
