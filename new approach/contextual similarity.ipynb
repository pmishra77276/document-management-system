```python
for i in paths:
    name=i.split('/')[-1].split('.')[1]
    print("Processing",name,"Document")
    with open(i,'r') as file:
        new_context=file.read()
    docs=text_splitter.split_text(new_context)
    docs=model.embed_documents(docs)
    doc=np.zeros(len(docs[0]))
    for j in docs:
        doc+=j
    emb_dict[i]=doc
```
